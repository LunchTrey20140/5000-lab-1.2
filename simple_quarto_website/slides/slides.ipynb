```python
#| echo: true
#| code-fold: true
#| code-summary: "Code"
#| code-overflow: Scroll

import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset("penguins")

# Set theme
sns.set_theme(style = 'white', palette = "Pastel1")

# Assign an object to the .relplot command (based on the plot, it's a side by side scatterplot categorized by island across the plots and species within them)
g = sns.relplot(
    data = df,
    x = "bill_length_mm",
    y = "flipper_length_mm",
    hue = "species",
    col = "island",
    s = 200)

# Print Types
print('-----TYPES-----')
print(type(g.fig))
print(type(g.axes[0,0]))
print(type(g.axes[0,1]))

# Font Size
FS = 18

# Axis Titles 
g.axes[0,0].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,1].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,2].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,0].set_ylabel('Penguin flipper length (mm)', fontsize=FS)

# Axis Tic Font Size
for i in range(0,2):
    g.axes[0,i].tick_params(axis='both', which='major', labelsize=FS)

# Titles of each island's scatterplot
g.axes[0,0].set_title("Torgersen Island",fontsize=FS)
g.axes[0,1].set_title("Biscoe Island",fontsize=FS)
g.axes[0,2].set_title("Dream Island", fontsize=FS)

# Modify the Legend
plt.setp(g._legend.get_texts(), fontsize=FS)
plt.setp(g._legend.get_title(), fontsize=FS)
```
